### Full code

```python
# Creating Driver
driver = webdriver.Chrome()

# Go to the characters in books page
page_url = "https://witcher.fandom.com/wiki/Category:Characters_in_the_stories"
driver.get(page_url)

#Find books by categories
book_categories = driver.find_elements(By.CLASS_NAME, 'category-page__member-link')

books = []
for category in book_categories:
    book_url = category.get_attribute('href')
    book_name = category.text
    books.append({'book_name': book_name, 'url' : book_url})
```

```python
books
```

```python
character_list = []
for book in books:
    #go to book page
    driver.get(book['url'])

    book_characters = driver.find_elements(By.CLASS_NAME, 'category-page__member-link')

    for characters in book_characters:
        character_list.append({'book': book['book_name'], 'character':characters.text})
```

```python
character_list
```

```python
pd.DataFrame(character_list)
```

```python
excel_file = 'witcher_characters.xlsx'  # File path to save Excel file
df = pd.DataFrame(character_list)
df.to_excel(excel_file, index=False)  # index=False to exclude row numbers in Excel
```
